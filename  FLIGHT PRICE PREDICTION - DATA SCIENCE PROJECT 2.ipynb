# FEATURE ENGINEERING

```python
big['Date']=big['Date_of_Journey'].str.split('/').str[0]
big['Month']=big['Date_of_Journey'].str.split('/').str[1]
big['Year']=big['Date_of_Journey'].str.split('/').str[2]
big.head()
```

```python
big.dtypes
```

```python
big['Date']=big['Date'].astype(int)
big['Month']=big['Month'].astype(int)
big['Year']=big['Year'].astype(int)
big.dtypes
```

```python
big=big.drop(['Date_of_Journey'],axis=1)
big.head()
```

```python
big['Arrival_Time']=big['Arrival_Time'].str.split(' ').str[0]
big.head()
```

```python
big[big['Total_Stops'].isnull()]
big['Total_Stops']=big['Total_Stops'].fillna('1 stop')
big['Total_Stops']=big['Total_Stops'].replace('non-stop','0 stop')
big.head()
```

```python
big['Stop'] = big['Total_Stops'].str.split(' ').str[0]
big.head()
```

```python
big['Stop']=big['Stop'].astype(int)
big=big.drop(['Total_Stops'],axis=1)
big.head()
```

```python
big['Arrival_Hour'] = big['Arrival_Time'] .str.split(':').str[0]
big['Arrival_Minute'] = big['Arrival_Time'] .str.split(':').str[1]
big['Arrival_Hour']=big['Arrival_Hour'].astype(int)
big['Arrival_Minute']=big['Arrival_Minute'].astype(int)
big=big.drop(['Arrival_Time'],axis=1)
big.head()
```

```python
big['Departure_Hour'] = big['Dep_Time'] .str.split(':').str[0]
big['Departure_Minute'] = big['Dep_Time'] .str.split(':').str[1]
big['Departure_Hour']=big['Departure_Hour'].astype(int)
big['Departure_Minute']=big['Departure_Minute'].astype(int)
big=big.drop(['Dep_Time'],axis=1)
big.head()
```

```python
big['Route_1']=big['Route'].str.split('→ ').str[0]
big['Route_2']=big['Route'].str.split('→ ').str[1]
big['Route_3']=big['Route'].str.split('→ ').str[2]
big['Route_4']=big['Route'].str.split('→ ').str[3]
big['Route_5']=big['Route'].str.split('→ ').str[4]
big.head()
```

```python
big['Price'].fillna((big['Price'].mean()),inplace=True)
big['Route_1'].fillna("None",inplace=True)
big['Route_2'].fillna("None",inplace=True)
big['Route_3'].fillna("None",inplace=True)
big['Route_4'].fillna("None",inplace=True)
big['Route_5'].fillna("None",inplace=True)
big.head()
```

```python
big=big.drop(['Route'],axis=1)
big=big.drop(['Duration'],axis=1)
big.head()
```

```python
big.isnull().sum()
```

```python
encoder=LabelEncoder()
big["Airline"]=encoder.fit_transform(big['Airline'])
big["Source"]=encoder.fit_transform(big['Source'])
big["Destination"]=encoder.fit_transform(big['Destination'])
big["Additional_Info"]=encoder.fit_transform(big['Additional_Info'])
big["Route_1"]=encoder.fit_transform(big['Route_1'])
big["Route_2"]=encoder.fit_transform(big['Route_2'])
big["Route_3"]=encoder.fit_transform(big['Route_3'])
big["Route_4"]=encoder.fit_transform(big['Route_4'])
big["Route_5"]=encoder.fit_transform(big['Route_5'])
big.head()
```

#  RANDOM FOREST REGRESSOR

```python
from sklearn.model_selection import RandomizedSearchCV
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 50, cv = 5, verbose=2, random_state=42, n_jobs = 1)
rf_random.fit(X_train,y_train)
```

```python
y_pred=rf_random.predict(X_test)
```

```python
import seaborn as sns

sns.distplot(y_test-y_pred)
```
